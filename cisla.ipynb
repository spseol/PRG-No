### Bitové operace s celými čísly

operátor | operace
:---|:---
<code>&#124;</code> | disjunkce -- OR
`&` | konjunkce -- AND
`^` | nonekvivalence -- XOR
`~` | negace -- NOT
`>>`&nbsp;&nbsp; | bitový posun doleva
`<<` | bitový posun doprava


```python
print('{0:06b} {0:6d}\n{1:06b} {1:6d}'.format(26,6))
```

```python
print('{0:06b} {0:6d}'.format( 26 | 6 ))
```

```python
print('{0:06b} {0:6d}'.format( 26 & 6 ))
```

```python
print('{0:06b} {0:6d}'.format( 26 ^ 6 ))
```

```python
print ('{0:06b} {0:6d}'.format( 6 << 2 ))
```

```python
print ('{0:06b} {0:6d}'.format( 26 >> 2 ))
```

```python
### pravda, nepravda -- bool()

Proměnná `True`, `False` jsou jen synonymem pro `1` a
`0`. Dále platí, že `False` je jen prázdný řetězec, `None` nebo číslo `0`. Vše ostatní
je True.

[[!code -NR python
>>> bool('-1234')
True
>>> bool('0')
True
>>> bool(0)
False
>>> bool(1)
True
>>> bool(7654)
True
>>> bool(776.4)
True
>>> True+True+True+True
4
>>> True+2
3
>>> 87534 * False
0
]]()

Pravdivostní výrazy mohou používat
: `==`, `!=` -- rovno/ nerovno
: `>` , `>=`, `<` ,`<=` -- větší/menší/nebo rovno
: `in` -- test příslušnosti do seznamu.


[[!code -NR python
>>> 22 == 22
True
>>> 22 != 22
False
>>> 22 != 15.23
True
>>> 'a' in 'ahoj'
True
>>> 'a' in 'Hele'
False
>>> 'ab' in 'ahoj'
False
>>> 8 in (1,2,8,16,32)
True
>>> 
]]()

V pravdivostních výrazech lze používat logické operace **and**, **or** a **not**.

```
